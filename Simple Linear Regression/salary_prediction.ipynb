```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0,10.0)

df = pd.read_csv('Salary_Data.csv')
df.shape
```

```python
df.head()
```

```python
X = df['YearsExperience'].values
Y = df['Salary'].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)

n = len(X)

numer = 0
denum = 0
for i in range(n):
    numer += (X[i]-mean_x) * (Y[i]-mean_y)
    denum += (X[i]-mean_x) ** 2
m = numer / denum
c = mean_y - (m * mean_x)

print(m,c)
```

```python
max_x = np.max(X) + 100
min_x = np.min(X) - 100

x = np.linspace(min_x,max_x,1000)
y = c + m * x

plt.plot(x,y,color='#58b970',label='Regression line')
plt.scatter(X,Y,c='#ef5423',label='Scatter plot')
plt.xlabel('Years Experience')
plt.ylabel('Expected Salary')
plt.legend
plt.show
```

```python
ss_t = 0
ss_r = 0
for i in range(n):
    y_pred = c + m * X[i]
    ss_t += (Y[i] - mean_y ) ** 2
    ss_r += (Y[i] - y_pred) ** 2
    
r2 = 1 - (ss_r/ss_t)
print(r2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
X = X.reshape((n,1))
lr = LinearRegression()
lr = lr.fit(X,Y)
Y_pred = lr.predict(X)
r2_score = lr.score(X,Y)
print(r2_score)
```
